Question 1: What is hypothesis testing in statistics?  

Ans.   Hypothesis testing is a fundamental concept in statistics used to make decisions or inferences about a population based on sample data. It helps determine whether there is enough statistical evidence to support a certain belief or claim (called a hypothesis) about a population parameter.

 Key Concepts:

Null Hypothesis (H‚ÇÄ):
This represents the default or no-effect assumption.
Example: ‚ÄúThere is no difference between the means of two groups.‚Äù

Alternative Hypothesis (H‚ÇÅ or Ha):
This represents what we aim to prove.
Example: ‚ÄúThere is a significant difference between the means of two groups.‚Äù

Test Statistic:
A numerical value calculated from the sample data (e.g., t, z, œá¬≤) used to decide whether to reject H‚ÇÄ.

Significance Level (Œ±):
The probability of rejecting the null hypothesis when it is actually true (commonly Œ± = 0.05).

P-value:
The probability of obtaining a result at least as extreme as the observed one, assuming H‚ÇÄ is true.

If p ‚â§ Œ±, reject H‚ÇÄ (evidence supports H‚ÇÅ).

If p > Œ±, fail to reject H‚ÇÄ (not enough evidence against H‚ÇÄ).

 Steps in Hypothesis Testing:

State the hypotheses:

H‚ÇÄ: The default assumption

H‚ÇÅ: The claim or effect you want to test

Choose the significance level (Œ±)

Select the appropriate test (e.g., t-test, z-test, chi-square test)

Compute the test statistic and p-value

Make a decision:

Reject H‚ÇÄ if p ‚â§ Œ±

Fail to reject H‚ÇÄ if p > Œ±

 Example:

Suppose a company claims that the average lifetime of their light bulbs is 1000 hours.
You collect a random sample and find an average of 980 hours.

H‚ÇÄ: Œº = 1000

H‚ÇÅ: Œº ‚â† 1000

Perform a t-test, calculate the p-value.

If p < 0.05 ‚Üí Reject H‚ÇÄ ‚Üí There is evidence that the average lifetime is not 1000 hours.


Question 2: What is the null hypothesis, and how does it differ from the alternative
hypothesis?

Ans.  In hypothesis testing, we begin with two opposing statements about a population parameter:
the null hypothesis (H‚ÇÄ) and the alternative hypothesis (H‚ÇÅ or Ha).

 1. Null Hypothesis (H‚ÇÄ):

The null hypothesis represents the default, no-change, or no-effect assumption.

It assumes that any difference or relationship observed in the data is due to random chance.

It is the statement we test against.

Example:
H‚ÇÄ: The average height of students = 170 cm.
(This means there is no difference from the assumed average.)

 2. Alternative Hypothesis (H‚ÇÅ or Ha):

The alternative hypothesis represents the claim, effect, or difference that the researcher wants to prove.

It suggests that something significant or different is happening in the data.

Example:
H‚ÇÅ: The average height of students ‚â† 170 cm.
(This means there is a difference from the assumed average.)

 3. Key Differences:
Feature	Null Hypothesis (H‚ÇÄ)	Alternative Hypothesis (H‚ÇÅ or Ha)
Meaning	Assumes no effect or no difference	Assumes there is an effect or difference
Purpose	Acts as the statement to be tested	Represents what the researcher wants to prove
Symbol	H‚ÇÄ	H‚ÇÅ or Ha
Example (two-tailed)	Œº = 50	Œº ‚â† 50
Example (one-tailed)	Œº ‚â§ 50	Œº > 50 (right-tailed)
 4. Decision Rule:

If the p-value ‚â§ Œ± (significance level) ‚Üí Reject H‚ÇÄ (evidence supports H‚ÇÅ)

If the p-value > Œ± ‚Üí Fail to reject H‚ÇÄ (not enough evidence to support H‚ÇÅ)


Question 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.

Ans.The significance level, denoted by Œ± (alpha), is a crucial concept in hypothesis testing. It represents the threshold for making a decision about whether to reject the null hypothesis (H‚ÇÄ).

 Definition:

The significance level (Œ±) is the probability of rejecting the null hypothesis when it is actually true.
In other words, it measures the risk of making a Type I error (false positive).

Typical values of Œ± are:

0.05 (5%) ‚Üí Most common

0.01 (1%) ‚Üí More stringent

0.10 (10%) ‚Üí More lenient

 Role in Hypothesis Testing:

Set before testing:
The researcher chooses Œ± before performing the test to define how much evidence is required to reject H‚ÇÄ.

Compare with p-value:
After calculating the p-value, the decision rule is:

If p ‚â§ Œ±, ‚Üí Reject H‚ÇÄ (evidence supports H‚ÇÅ)

If p > Œ±, ‚Üí Fail to reject H‚ÇÄ (not enough evidence against H‚ÇÄ)

Controls confidence level:
The confidence level = (1 ‚àí Œ±).

For Œ± = 0.05, confidence level = 95%
This means we are 95% confident our decision is correct.

 Example:

A manufacturer claims that the average life of a battery is 100 hours.
You test this claim with Œ± = 0.05.

H‚ÇÄ: Œº = 100

H‚ÇÅ: Œº ‚â† 100

After conducting the test, you get p = 0.03.

‚û° Since p (0.03) < Œ± (0.05),
you reject H‚ÇÄ, concluding that the battery life is significantly different from 100 hours.

 Interpretation of Œ±:
Œ± Value	Confidence Level	Interpretation
0.10	90%	Moderate evidence needed to reject H‚ÇÄ
0.05	95%	Standard level of evidence
0.01	99%	Strong evidence required to reject H‚ÇÄ

Question 4: What are Type I and Type II errors? Give examples of each.


Ans.  In hypothesis testing, we make decisions based on sample data ‚Äî but because samples are subject to chance, there is always a possibility of making errors in our conclusions.
The two types of possible errors are Type I and Type II errors.

 1. Type I Error (False Positive)

A Type I error occurs when we reject the null hypothesis (H‚ÇÄ) even though it is actually true.

In other words, we detect an effect or difference that does not really exist.

The probability of making a Type I error is represented by the significance level (Œ±).

Example:
Suppose a company claims a medicine has no side effects.

H‚ÇÄ: The medicine has no side effects.

H‚ÇÅ: The medicine has side effects.
If the test concludes that the medicine has side effects when it actually does not, that‚Äôs a Type I error.

‚Üí Analogy: A ‚Äúfalse alarm‚Äù ‚Äî you think there‚Äôs an effect when there isn‚Äôt.

 2. Type II Error (False Negative)

A Type II error occurs when we fail to reject the null hypothesis (H‚ÇÄ) even though it is false.

This means we miss a real effect or difference.

The probability of making a Type II error is represented by Œ≤ (beta).

The complement (1 ‚àí Œ≤) is called the power of the test, which measures the ability to detect a true effect.

Example:
Using the same medicine study:

H‚ÇÄ: The medicine has no side effects.

H‚ÇÅ: The medicine has side effects.
If the test concludes that the medicine has no side effects when it actually does, that‚Äôs a Type II error.

‚Üí Analogy: A ‚Äúmissed detection‚Äù ‚Äî you fail to notice a real effect.

 Comparison Table:
Error Type	Decision	Reality	Description	Symbol
Type I Error	Reject H‚ÇÄ	H‚ÇÄ is true	False positive (detecting a false effect)	Œ±
Type II Error	Fail to reject H‚ÇÄ	H‚ÇÄ is false	False negative (missing a real effect)	Œ≤
 Example in Everyday Context:

Court Trial Analogy:

H‚ÇÄ: The defendant is innocent.

H‚ÇÅ: The defendant is guilty.

Outcome	Type of Error	Interpretation
Convict an innocent person	Type I Error	Rejecting H‚ÇÄ when true
Acquit a guilty person	Type II Error	Failing to reject H‚ÇÄ when false

Question 6: Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.
(Include your Python code and output in the code box below.)
Hint: Generate random number using random function.


Ans.  



```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10      # number of trials
p = 0.5     # probability of success
size = 1000 # number of random samples

# Generate random numbers using the binomial distribution
data = np.random.binomial(n, p, size)

# Display the generated data‚Äôs mean and standard deviation
print("Mean:", np.mean(data))
print("Standard Deviation:", np.std(data))

# Plot the histogram
plt.hist(data, bins=range(n + 2), edgecolor='black', align='left')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.show()
```

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9].



Ans.  

```python
# Import required libraries
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesis setup
mu_population = 50      # Claimed population mean
sigma_population = 0.5  # Known population standard deviation
alpha = 0.05            # Significance level

# Step 1: Calculate sample statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Step 2: Compute the Z-statistic
z_statistic = (sample_mean - mu_population) / (sigma_population / np.sqrt(n))

# Step 3: Compute p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

# Display results
print("Sample Mean:", round(sample_mean, 3))
print("Z-statistic:", round(z_statistic, 3))
print("P-value:", round(p_value, 3))

# Step 4: Decision
if p_value < alpha:
    print("‚úÖ Reject the null hypothesis (H‚ÇÄ).")
else:
    print("‚ùå Fail to reject the null hypothesis (H‚ÇÄ).")
```

Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.  



Ans.  

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Generate data from a normal distribution
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=100, scale=15, size=200)  # mean=100, std=15, n=200

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)
standard_error = sample_std / np.sqrt(n)

# Step 3: Compute 95% confidence interval
confidence_level = 0.95
z_critical = stats.norm.ppf((1 + confidence_level) / 2)  # Z = 1.96 for 95%
margin_of_error = z_critical * standard_error
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Step 4: Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")

# Step 5: Plot histogram and confidence interval
plt.hist(data, bins=20, color='lightblue', edgecolor='black')
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Mean = {sample_mean:.2f}')
plt.axvline(confidence_interval[0], color='green', linestyle='--', label=f'Lower 95% CI = {confidence_interval[0]:.2f}')
plt.axvline(confidence_interval[1], color='green', linestyle='--', label=f'Upper 95% CI = {confidence_interval[1]:.2f}')
plt.title('Normal Distribution with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.



Ans.   

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    """
    Calculate Z-scores for a given dataset.
    Formula: Z = (X - mean) / standard_deviation
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores, mean, std_dev

# Generate random dataset
data = np.random.normal(70, 15, 1000)  # mean=70, std=15, 1000 samples

# Calculate Z-scores
z_scores, mean, std_dev = calculate_z_scores(data)

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(z_scores, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram of Z-scores (Standardized Data)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color='red', linestyle='--', linewidth=1.5, label='Mean (Z = 0)')
plt.legend()
plt.show()

# Show first few results
z_scores[:10], mean, std_dev
```
